```python
from osgeo import gdal
import matplotlib.pyplot as plt
import os
```

```python
sent_path = "res/sent1-march-v1/"
smap_path = "res/smapL4-march-v1/"
base_path = 'res/result-out/'
```

```python
s_paths = os.listdir(sent_path)
```

```python
def create_dir(name,base_path = 'res/result-out/'):
    path = base_path + name + "/"
    if not os.path.exists(path):
        print(f"'{path}' doesn't existed before", "Creating directory...")
        os.makedirs(path)
        print(f"Successfully Created '{path}'")
    t = os.listdir(path)
    l = len(t)
    new_path = name + '-'+str(l)
    print(new_path)
    dir_path = path+new_path + '/'
    try:
        os.makedirs(dir_path)
    except :
        print(f"error occurred \nwhile creating directory at \n-->'{dir_path}'")
    return (l, dir_path)
    
```

```python
def visualize_data(path, save_file = False,name = "default"):
    if not os.path.exists(path):
        print("Invalid Path")
        return
    dir_path = ""
    if save_file:
        l, dir_path = create_dir(name)
    paths = os.listdir(path)
    for p in paths:
        print(p)
        sp = path + p;
        g = gdal.Open(sp)
        im = g.ReadAsArray()
        plt.imshow(im)
        plt.show()
        if not save_file:
            print("Error: Required parameter `dir_path` to save file.")
            continue
        sp_path = dir_path + p.split('.')[0] + ".png"
        plt.imsave(sp_path,im)
        
```

```python
# l, dir_path = create_dir('sent')
visualize_data(sent_path, name = 'sent', save_file = True)
```

```python

visualize_data(smap_path, name = 'smap', save_file = True)
```
